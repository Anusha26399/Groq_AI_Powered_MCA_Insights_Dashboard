```python
import pandas as pd

# Load the combined MCA dataset
df = pd.read_csv("MCA_Data_Cleaned.csv", low_memory=False)

# Clean column names
df.columns = df.columns.str.strip().str.replace(" ", "_")

# Define target states
target_states = ["maharashtra", "gujarat", "delhi", "tamil nadu", "karnataka"]

# Filter the dataset
df_assignment = df[df["CompanyStateCode"].str.lower().isin(target_states)]

# Remove duplicates by CIN
df_assignment = df_assignment.drop_duplicates(subset="CIN", keep="last")

# Optional: Reset index
df_assignment = df_assignment.reset_index(drop=True)

# Save filtered master dataset
df_assignment.to_csv("MCA_Master_Assignment.csv", index=False)

# Quick checks
print("‚úÖ Filtered master dataset created!")
print("Shape:", df_assignment.shape)
print("\nüìç Companies per state:\n", df_assignment["CompanyStateCode"].value_counts())
print("\nüè¢ Distinct ROC codes:\n", df_assignment["CompanyROCcode"].unique())
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

# -------------------------------
# Step 1: Load your master dataset (Day 1)
# -------------------------------
df_day1 = pd.read_csv("MCA_Master_Assignment.csv")

# -------------------------------
# Helper function to simulate changes
# -------------------------------
def simulate_day(df_prev, new_companies_count=5, remove_count=5, authcap_frac=0.01, status_frac=0.005):
    df_new = df_prev.copy()
    
    # 1. Randomly update AuthorizedCapital
    update_idx = df_new.sample(frac=authcap_frac, random_state=np.random.randint(0,1000)).index
    df_new.loc[update_idx, 'AuthorizedCapital'] = df_new.loc[update_idx, 'AuthorizedCapital'] * np.random.uniform(1.1, 1.5, len(update_idx))
    
    # 2. Randomly update CompanyStatus
    status_idx = df_new.sample(frac=status_frac, random_state=np.random.randint(0,1000)).index
    df_new.loc[status_idx, 'CompanyStatus'] = np.where(df_new.loc[status_idx, 'CompanyStatus']=='Active', 'Strike Off', 'Active')
    
    # 3. Add new companies
    new_companies = df_new.sample(new_companies_count, random_state=np.random.randint(0,1000)).copy()
    new_companies['CIN'] = ['NEWCIN' + str(np.random.randint(10000,99999)) for _ in range(new_companies_count)]
    new_companies['CompanyRegistrationdate_date'] = datetime.today().strftime('%d-%m-%Y')
    df_new = pd.concat([df_new, new_companies], ignore_index=True)
    
    # 4. Remove companies
    remove_idx = df_new.sample(remove_count, random_state=np.random.randint(0,1000)).index
    df_new = df_new.drop(remove_idx).reset_index(drop=True)
    
    return df_new

```

```python
# -------------------------------
# Step 2: Generate Day 2 and Day 3 snapshots
# -------------------------------
df_day2 = simulate_day(df_day1)
df_day3 = simulate_day(df_day2)

# Save snapshots
df_day1.to_csv("MCA_Day1.csv", index=False)
df_day2.to_csv("MCA_Day2.csv", index=False)
df_day3.to_csv("MCA_Day3.csv", index=False)
```

```python
# -------------------------------
# Step 3: Function to generate daily change logs
# -------------------------------
def generate_change_log_fast(df_prev, df_new, log_filename):
    # 1. Identify new and removed companies
    removed = df_prev.loc[~df_prev['CIN'].isin(df_new['CIN']), ['CIN']]
    removed['Change_Type'] = 'Deregistration'
    removed['Field_Changed'] = None
    removed['Old_Value'] = None
    removed['New_Value'] = None

    new = df_new.loc[~df_new['CIN'].isin(df_prev['CIN']), ['CIN']]
    new['Change_Type'] = 'New Incorporation'
    new['Field_Changed'] = None
    new['Old_Value'] = None
    new['New_Value'] = None

    # 2. Compare common companies using merge
    common = df_prev.merge(df_new, on='CIN', suffixes=('_old', '_new'))
    diff_records = []

    cols_to_check = ['AuthorizedCapital', 'PaidupCapital', 'CompanyStatus', 'CompanyClass']

    for col in cols_to_check:
        changed = common[common[f'{col}_old'] != common[f'{col}_new']]
        if not changed.empty:
            temp = changed[['CIN', f'{col}_old', f'{col}_new']].copy()
            temp['Change_Type'] = 'Field Update'
            temp['Field_Changed'] = col
            temp.rename(columns={f'{col}_old': 'Old_Value', f'{col}_new': 'New_Value'}, inplace=True)
            diff_records.append(temp[['CIN', 'Change_Type', 'Field_Changed', 'Old_Value', 'New_Value']])

    diffs = pd.concat([removed, new] + diff_records, ignore_index=True)
    diffs.to_csv(log_filename, index=False)

    print(f"‚úÖ Change log saved: {log_filename}, Total changes: {len(diffs)}")
```

```python
# -------------------------------
# Step 4: Generate change logs
# -------------------------------
generate_change_log_fast(df_day1, df_day2, "Daily_Change_Log_Day2.csv")
generate_change_log_fast(df_day2, df_day3, "Daily_Change_Log_Day3.csv")
```
